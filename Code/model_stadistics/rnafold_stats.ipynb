```python
import os
import csv
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python

archive = "RNAFOLD_result_stats.csv"

path = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
stats_path = (path + "/Code/saved_stats/" + archive)

list_bin_acc = []
list_bin_recall = []
list_bin_f = []
list_bin_prec = []
list_ex_acc = []
list_ex_recall = []
list_ex_f = []
list_ex_prec = []
list_ex_amm = []

with open(stats_path, newline='') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter='\t')
    data = list(csv_reader)

elements = list(map(float,data[0][1:-1]))

list_bin_acc = list(map(float,data[1][1:]))
list_bin_recall = list(map(float,data[2][1:]))
list_bin_f = list(map(float,data[3][1:]))
list_bin_prec = list(map(float,data[4][1:]))

list_ex_acc = list(map(float,data[5][1:]))
list_ex_recall = list(map(float,data[6][1:]))
list_ex_f = list(map(float,data[7][1:]))
list_ex_prec = list(map(float,data[8][1:]))
list_ex_amm = list(map(float,data[9][1:]))
for i in range(len(list_ex_amm)):
    list_ex_amm[i] = list_ex_amm[i]/100
```

```python
# plotting the data
figure(figsize=(20, 3), dpi=80)
plt.plot(elements, list_bin_acc)
# Adding the title
plt.title("Binary accuracy ")
# Adding the labels
plt.ylabel("Accuracy")
plt.xlabel("Samples")
plt.show()
# plotting the data
figure(figsize=(20, 3), dpi=80)
plt.plot(elements, list_bin_recall)
# Adding the title
plt.title("Binary recall ")
# Adding the labels
plt.ylabel("Recall")
plt.xlabel("Samples")
plt.show()
# plotting the data
figure(figsize=(20, 3), dpi=80)
plt.plot(elements, list_bin_f)
# Adding the title
plt.title("Binary F1 ")
# Adding the labels
plt.ylabel("F1")
plt.xlabel("Samples")
plt.show()
# plotting the data
figure(figsize=(20, 3), dpi=80)
plt.plot(elements, list_bin_recall)
# Adding the title
plt.title("Binary recall ")
# Adding the labels
plt.ylabel("Recall")
plt.xlabel("Samples")
plt.show()
# plotting the data
figure(figsize=(20, 3), dpi=80)
plt.plot(elements, list_bin_prec)
# Adding the title
plt.title("Binary Precision ")
# Adding the labels
plt.ylabel("Precision")
plt.xlabel("Samples")
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(20,3))
# plotting the data in the 1st subplot
axes.plot(elements, list_ex_acc)
# plotting the data in the 1st subplot only
axes.plot(elements, list_ex_amm)
plt.title("Element Accuracy (blue) and the rate of perfect matches (orange). ")
plt.ylabel("Accuracy")
plt.xlabel("Samples")
plt.show()
```
